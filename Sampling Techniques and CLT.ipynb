### 9. Drawing a Stratified Random Sample of size 400 based on the Variable Job Level and plotting a histogram      for the Sample

**Stratified random sampling** is a method of sampling that involves the division of a population into smaller sub-groups known as **strata**. 

                                            (or)
                                            
A stratified random sampling involves dividing the entire population into homogeneous groups called strata. Random samples are then selected from each stratum.


In stratified random sampling, or stratification, the strata are formed based on members' shared attributes or characteristics such as job level, income, gender, etc.

Stratified sampling is used to highlight differences between groups in a population, as opposed to simple random sampling, which treats all members of a population as equal, with an equal likelihood of being sampled.

Here we are checking how the data is distributed based on job levels so that we can get a clear view of the data

```python
#Checking how the data is distributed for various job levels in the dataset
df.groupby('JobLevel').count()
```

```python
#Creating an empty datframe to store the Stratified Sample data
master_strata = pd.DataFrame()
for  level in range(1,6):
    
    #Creating a population strata based on monthly income for each job level
    pop_strata = df[df.JobLevel == level][['MonthlyIncome','JobLevel']]
    
    #If the length of the population strata is > 80, then we no need to replace the data and we can generate the sample data.
    if len(pop_strata) > 80:
    
        sample_strata = pop_strata.sample(n = 80, random_state = 50)
        
    #If the length of the population strata is < 80, then we need to replace the data and we can generate the sample data.
    else:
        sample_strata = pop_strata.sample(n = 80, replace = True,  random_state = 50)
    
    #Concatinating the emppty data frame and sample strata we created.
    master_strata = pd.concat([master_strata, sample_strata])

#Printing the strata sizes to get an overview
print('The Strata Sizes:\n', master_strata.groupby('JobLevel').count())
```

```python
sample_1 = data.sample(n = 400, random_state=40)
#Plotting the histogram of the above sample 
plt.figure(figsize=(7,4))
plt.hist(sample_1.MonthlyIncome, edgecolor = 'black', color = 'skyblue')
plt.xlabel('Monthly Income of the sample')
plt.ylabel('Frequency')
plt.title('Histogram of Monthly Income of the Sample 1 of size 400')
plt.show()

#Plotting the histogram of the above sample 
plt.figure(figsize=(7,4))
plt.hist(master_strata.MonthlyIncome, edgecolor = 'black', color = 'gold')
plt.xlabel('Monthly Income of the Stratified sample')
plt.ylabel('Frequency')
plt.title('Histogram of Monthly Income of the Stratified Random Sample of size 400')
plt.show()
```

The Random sample produced a right skewed but the Stratified random sample is more **smoother**.

The random sample looks similar to the population but the stratified random sample is not looking similar to the population.

This is because the **Job Levels are skewed towards lower levels and the level 4 and 5 are less.**

## Part B: Sampling Distribution

### 1. Generating 1000 samples of size 400 and plotting the histogram of this 1000 sample means

```python
samp_means = list()
for i in range(1,1001):
    
    sample = data.sample(n = 400)
    
    samp_means.append(sample.MonthlyIncome.mean())

#Setting the figure size
plt.figure(figsize=(7,4))

#Plotting the histogram of the above sample
plt.hist(samp_means, edgecolor = 'black', color = 'green')

#Label for the x axis
plt.xlabel('Monthly Income of the Samples')
    
#Label for the y axis
plt.ylabel('Frequency')
    
#Title of the Histogram
plt.title('Histogram of Monthly Income of the 1000 Sample Means ') 
plt.show()
```

When the sample means distribution is plotted, the histogram resembles that of a **normal distribution**

### 2. Calculating the standard Error of the above sample mean

**Standard Error** of a statistic is the standard deviation of it's sampling distribution.. or an estimate of the standard deviation..

If the statistic is sample mean, then it is called as **Standard Error of the Sample Mean**

```python
standard_error = np.std(samp_means)

print('Standard Error: ', standard_error)
```

### 3. Checking whether CLT holds in this sample

CLT means Central Limit Theorem

```python
#Printing the Standard Error of the Sample 
print('\nStandard Error: ', standard_error)

#Calculating the population standard deviation
population_std = data.MonthlyIncome.std()

#Calculating the population standard error
population_std= population_std/np.sqrt(400)
print('\nPopulation Standard Error/Sqrt(n)', population_std)

#Printing the sampling distribution mean
print('\nSampling Distribution Mean:', np.mean(samp_means))

#Printing the population mean
print('\nPopulation Mean', data.MonthlyIncome.mean() )
```

**From the Descriptive Statistics could see that the Mean of the sampling distribution is approxiamtely equal to the Mean of the population Mean. Also, the sigma/Sqrt(n) is nearly equal to the S.D of the Sampling Distribution. Also, the Sampling distribution seems to be normal. SO we can say that the CLT is hold up in this case.**

### 4. Calculating 95% Confidence level Interval of mean from each samples and checking how many of these CI intervals captures the true population mean

```python
#Importing the required library
import scipy.stats as stats

#Calculating the population standard deviation
population_std = data.MonthlyIncome.std()

#Calculating the population mean
population_mean = data.MonthlyIncome.mean() 

# As the CI is 95%, the alpha is 0.05. So, alpha/ 2 is 0.025

# ZScore for pdf of 0.975
z_score = stats.norm.ppf(0.975)

# no of samples taken
no_samples = 1000

# sample size
n = 400

# counter for counting the no of samples means which contain the true population mean
in_range_count = 0

for i in range(0,no_samples):
    
    sample = data.sample(n = n)
    
    sample_mean = sample.MonthlyIncome.mean()
    
    lower_ci = sample_mean - z_score*population_std/np.sqrt(n)
    upper_ci = sample_mean + z_score*population_std/np.sqrt(n)
    
    
    if population_mean >= lower_ci and population_mean <= upper_ci:
        
        in_range_count += 1  
  
print('Percentage of Intervals that contain the Population Mean:',in_range_count*100/no_samples)
```

From the CI given as 95%, the expectation is that 95% of the sample's Confidence Interval will contain the true population mean.

But due to the random sampling process, it fluctates around 95%.

Here could see that for **98.4 percent** of the Confidence intervals formed contains the true population mean.

### 6) Show that on average a bootstrap sample contain 63% (approx.) of the original data in the original sample.

```python
# list that will contain sample count that are common
sample_counts= list()

# sample size
sample_size = 400

# taking sample from the population
sample_s = data.sample(n = sample_size, random_state=100)

# running the loop for 1000 times
for i in range(1,1001):
    
    # bootstrap sample
    b_sample = np.random.choice(sample_s.index.values, size =  sample_size, replace = True)
    
    # calculating the unique samples
    unique_samples = np.unique(b_sample)
    # append the count
    sample_counts.append(np.count_nonzero(unique_samples))

# calculating the average sample counts
mean_sample_count = np.mean(sample_counts)

# Changing it into percentage
print(mean_sample_count*100/sample_size)
```

**Could see that on an average bootstrap samples contain 63% of the original data, the others are replicating**
